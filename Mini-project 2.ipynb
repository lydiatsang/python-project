## 1. Objective
PIL (Python Image Library, or Python Imaging Library) is the fact standard for processing graphics on the Python platform, with powerful features and a concise API. The main objective of PIL is to provide a solid foundation for a general image processing tool. PIL dds image processing capabilities to our Python interpreter, and handles raster images(that is, rectangles of pixel data). The core image library is designed for fast access to data stored in a few basic pixel formats. Also, PIL provides extensive file format support, an efficient internal representation, and fairly powerful image and graphics processing capabilities. The problem it is trying to solve includes image archives, image display and image processing. For example, it can read, write, cut, paste, and merge images, do geometrical and color transforms, deal with image enhancement and sequences, print postscript, and control the decoder.

### Example 1: 
Here we want to use ImageEnhance module to enhance the color, brightness, contrast, and sharpness of original picture without changing the original picture. Instead, we will return a modifed copy to the folder.

```python
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFilter
from PIL import ImageEnhance
import random
import sys
# Load the image
imgU = Image.open("UtaPuri.jpg")

# Convert the jpg file to RGB
imgU = imgU.convert("RGB")              


# Try to use PIL lambda enhance brightness 
imgbrightU=imgU.point(lambda i : i*1.5) # Enhance the brightness of every pixels by 50%

# Save the image to local
imgbrightU.save("UtaPuriChange2.jpg")

# Try to use PIL to play with Color and save the image to local
im_5 = ImageEnhance.Color(imgU).enhance(0.5)
im_5.save("im_5.jpg")

im_20 =ImageEnhance.Color(imgU).enhance(2.0)
im_20.save("im_20.jpg")

im_30 = ImageEnhance.Color(imgU).enhance(3.0)
im_30.save("im_30.jpg")
```

```python
# Let's do several processing at one time

#PIL ImageEnhance
istep=4
irange=4.0

# Color
imgenhancer_Color=ImageEnhance.Color(imgU)
for i in range(istep):
        factor=i/irange
        img_enhance_color=imgenhancer_Color.enhance(factor)
        #img_enhance_color.show("Color %f" %factor)
        img_enhance_color.save("Color_%.2f.jpg" %factor) 

# Brightness
imgenhancer_Brightness=ImageEnhance.Brightness(imgU)
for i in range(istep):
        factor=i/irange
        img_enhance_Brightness=imgenhancer_Brightness.enhance(factor)
        #img_enhance_Brightness.show("Brightness %f" %factor)
        img_enhance_Brightness.save("Brightness_%.2f.jpg" %factor) 

# Contrast
imgenhancer_Contrast=ImageEnhance.Contrast(imgU)
for i in range(istep):
        factor=i/irange
        img_enhance_Contrast=imgenhancer_Contrast.enhance(factor)
        #img_enhance_Contrast.show("Contrast %f" %factor)
        img_enhance_Contrast.save("Contrast_%.2f.jpg" %factor) 
        
# Sharpness        
imgenhancer_Sharpness=ImageEnhance.Sharpness(imgU)
for i in range(istep):
        factor=i/irange
        img_enhance_Sharpness=imgenhancer_Sharpness.enhance(factor)
        #img_enhance_Sharpness.show("Sharpness %f" %factor)
        img_enhance_Sharpness.save("Sharpness_%.2f.jpg" %factor) 
#end
```

## 6. List other interesting or useful features and functionalities of the library (additional examples are not required)

There are several other useful funtion in PIL. 

1. You can use PIL to turn a photo into a sketch picture. To achieve this, first you need to turn the picture into L mode (img = img.convert("L")), and use function to get the Red, Green, Blue threshhold, and finally you will get the sketch.

2. You also use Image.paste(im, box=None, mask=None) to cover certain area over with mosaics. 

3. The point() method can be used to translate the pixel values of an image (e.g. image contrast manipulation). In most cases, a function object expecting one argument can be passed to this method. Each pixel is processed according to that function. So, we can use this method to process each pixels, and solve ML problems.

4. We can also use PIL to process simple GIF. PIL contains some basic support for image sequences (also called animation formats). Supported sequence formats include FLI/FLC, GIF, and a few experimental formats. TIFF files can also contain more than one frame. When you open a sequence file, PIL automatically loads the first frame in the sequence. You can use the seek and tell methods to move between different frames.

5. We can use PIL to generate random characters in random colors and random size. The ImageDraw function provides simple 2D graphics for Image objects. You can use this module to create new images, annotate or retouch existing images, and to generate graphics on the fly for web use.

Below is an example of feature 5 talked above.

```python
# create verification code
# create random character image
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
from PIL import ImageFilter

import random

# random character:
def rndChar():
    return chr(random.randint(65, 90))

# random color 1:
def rndColor():
    return (random.randint(64, 255), random.randint(64, 255), random.randint(64, 255))

# ramdom color 2:
def rndColor2():
    return (random.randint(32, 127), random.randint(32, 127), random.randint(32, 127))

# assign the image size:
width = 60 * 4
height = 60
image = Image.new('RGB', (width, height), (255, 255, 255))
# create font object:
font = ImageFont.truetype('arial.ttf', 36)
# create draw object:
draw = ImageDraw.Draw(image)
# fill in the image:
for x in range(width):
    for y in range(height):
        draw.point((x, y), fill=rndColor())
# print the output:
for t in range(4):
    draw.text((60 * t + 10, 10), rndChar(), font=font, fill=rndColor2())
    
# apply the blur function:
image = image.filter(ImageFilter.BLUR)
image.save('code.jpg', 'jpeg')
imshow(np.asarray(image))
```
